# Ответ на вопрос первого нулевого занятия

### Задание
Вопросы по заданию
В качестве первого задания мы предлагаем вам сформулировать учебные цели по критериям SMART.

Это упражнение не на проверку, оно поможет вам осознанно спланировать траекторию вашего обучения, корректировать ее или же детализировать сами цели.

Целей может быть несколько, но важно обозначить для себя, как вы поймете, что достигли ее.

Базовое определение критериев SMART.

Удачи!

### Решение 

#### Цель по SMART

К окончанию срока обучения на курсе Аналитика данных (в идеале до конца текущего календарного года) развить компетенции аналитика данных, повысить навыки программирования на Python через освоение дополнительных, специализированных библиотек, до уровня, необходимого для получения работы по новой квалификации, а также достаточной для возможности получения удаленной работы в одной из крупных сетевых ритейлов.
